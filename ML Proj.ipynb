```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(100, 2)
k = 3
centroids = data[np.random.choice(len(data), k, replace=False)]

for _ in range(10):
    distances = np.array([[np.linalg.norm(point - c) for c in centroids] for point in data])
    clusters = np.argmin(distances, axis=1)
    new_centroids = []
    for i in range(k):
        points_in_cluster = data[clusters == i]
        if len(points_in_cluster) > 0:
            new_centroids.append(points_in_cluster.mean(axis=0))
        else:
            new_centroids.append(np.random.rand(2))
    new_centroids = np.array(new_centroids)
    if np.allclose(centroids, new_centroids):
        break
    centroids = new_centroids

for i in range(k):
    cluster_points = data[clusters == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1])
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='*', s=100)
plt.show()
```
