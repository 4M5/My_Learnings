```python
#dfs
def dfs(graph,start,target):
  stack=[[start]]
  visited=set()
  while stack:
    path=stack.pop()
    node=path[-1]
    if node==target:
      return path
    if node not in visited:
      visited.add(node)
      for neighbor in graph[node]:
        new_path=list(path)
        print(new_path)
        new_path.append(neighbor)
        stack.append(new_path)
  return None
graph={
      1:[2,3],
      2:[1,4,5],
      3:[1,6],
      4:[2,7],
      5:[2,8],
      6:[3],
      7:[4],
      8:[5]
}
start=int(input("Enter start node: "))
target=int(input("Enter Target node: "))
path=dfs(graph,start,target)
print(f"Path from {start} to {target} using dfs: {path}")
```

```python
#Backtrack
regions=["WA", "NT", "Q", "NSW", "SA", "V", "T"]
colors=["Red", "Green", "Blue"]
neighbors = {
"WA": ["NT", "SA"],
"NT": ["WA", "SA", "Q"],
"Q": ["NT", "SA", "NSW"],
"NSW": ["SA", "Q", "V"],
"SA": ["WA", "NT", "Q", "NSW", "V"],
"V": ["SA", "NSW"],
"T": []
}
def is_consistent (region, color, assignment):
  for neighbor in neighbors [region]:
    if neighbor in assignment and assignment [neighbor]==color:
      return False
  return True
def backtrack (assignment):
  if len(assignment) == len(regions):
    return assignment
  for region in regions:
    if region not in assignment:
      break
  print(assignment)
  for color in colors:
    if is_consistent (region, color, assignment):
      assignment [region] = color
      result = backtrack (assignment)
      if result:
        return result
      del assignment [region]
  return None
def solve_map_coloring():
  assignment = {}
  return backtrack (assignment)
solution = solve_map_coloring()
if solution:
  print("Solution found: ")
  for region, color in solution.items():
    print(f" {region}: {color}")
else:
  print("No solution exists.")
```
