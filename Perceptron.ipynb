```python
#TRAINING-----------------
# 1. Load DATA
x=[[1,1],[4,15],[2,1]]
y=["Boy","Girl","Boy"]

# 2. Load MODEL
from sklearn.linear_model import Perceptron
model=Perceptron()

# 3. Fit DATA into MODEL
modell=model.fit(x,y)

#TESTING---------(predict)
op=modell.predict([[15,16],[0,0]])
print(op)
```
