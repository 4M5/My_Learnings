```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/content/iris.csv")
data.head()
x=data[["sepal_length","sepal_width","petal_length","petal_width"]]
x1=x[["sepal_length"]]
x2=x[["sepal_width"]]

from sklearn.cluster import DBSCAN
model=DBSCAN(eps=0.1,min_samples=5)
model2=model.fit(x)
labels1=model2.labels_
plt.scatter(x1,x2,c=labels1,cmap="rainbow")
```

```python
#HALF MOON
from sklearn.datasets import make_moons
w,_ = make_moons(n_samples=1000, noise=0.05)
w1=w[:,0]
w2=w[:,1]
model3=model.fit(w)
labels3=model3.labels_
plt.subplot(1,2,1)
plt.scatter(w1,w2,c=labels3,cmap="rainbow")
plt.subplot(1,2,2)
#HALF MOON
from sklearn.datasets import make_moons
w,_ = make_moons(n_samples=1000, noise=0)
w1=w[:,0]
w2=w[:,1]
model4=model.fit(w)
labels4=model4.labels_
plt.scatter(w1,w2,c=labels4,cmap="rainbow")
print(labels4)
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
np.random.seed(42)
data2=np.random.rand(10,4)#10row and 4col
linked1=linkage(data2,method="average")
linked2=linkage(data2,method="single")
linked3=linkage(data2,method="complete")
link=input("Choose link")
plt.figure(figsize=(10,6))
dendrogram(linked1,orientation='top',labels=range(1,11))
plt.title("Simple DEND")
plt.xlabel("Sample Index")
plt.ylabel("Dist")
```

```python
plt.figure(figsize=(10,6))
dendrogram(linked2,orientation='right',labels=range(1,11))
plt.title("Simple DEND")
plt.xlabel("Sample Index")
plt.ylabel("Dist")
```
