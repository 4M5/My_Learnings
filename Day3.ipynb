1. MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10])
y=np.array([0,500])
plt.scatter(x,y)
plt.plot(x,y)
```

```python
yp=np.array([3,8,1,10])
plt.plot(yp,marker='*',color='blue')
```

```python
#first subplot in 2x2 grid
plt.subplot(2,2,1)
plt.plot([10,5,8,1],[1,2,3,4])
plt.title("PLOT 1")

#second subplot in 2x2 grid
plt.subplot(2,2,2)
plt.plot([50,8,3,2],[6,5,4,3])
plt.title("PLOT 2")

#second subplot in 2x2 grid
plt.subplot(2,2,3)
plt.plot([5,8,6,2],[0,5,2,10])
plt.title("PLOT 3")

#second subplot in 1x2 grid
plt.subplot(2,2,4)
plt.plot([13,11,0,1],[1,5,5,10])
plt.title("PLOT 4")
```

```python
plt.subplot(2,2,1)
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)
plt.title("SINE")
plt.plot(x,y,color='red',linewidth=3)

plt.subplot(2,2,2)
x=np.linspace(0,2*np.pi,100)
z=np.cos(x)
plt.title("COSINE")
plt.plot(x,z,color='red',linewidth=3)
```
