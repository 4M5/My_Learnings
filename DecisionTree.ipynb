```python
from sklearn.datasets import load_iris
data=load_iris()
x=data.data
y=data.target

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model1=model.fit(x,y)
print(model1.predict([[3,4,5,6]]))

from sklearn import tree
iri=tree.export_graphviz(model1,out_file=None,feature_names=data.feature_names,class_names=data.target_names,filled=True,rounded=True)
import graphviz
graph = graphviz.Source(iri)
graph
```

```python
import pydot
from IPython.display import Image
(graph,)=pydot.graph_from_dot_data(iri)
graph.write_png("DECISION TREE.png")
Image("DECISION TREE.png")
```

```python
from sklearn.datasets import load_iris
data=load_iris()
x=data.data
y=data.target

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model1=model.fit(x,y)
print(model1.predict([[3,4,5,6]]))

from sklearn import tree
iri=tree.export_graphviz(model1,out_file=None,feature_names=data.feature_names,class_names=data.target_names,filled=True,rounded=True)
import graphviz
graph = graphviz.Source(iri)
graph
```
