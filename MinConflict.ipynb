```python
import random
def count_conflicts(queens,col,row):
  conflicts=0
  for c in range(len(queens)):
    if c==col:
      continue
    r=queens[c]
    if r==row or abs(c-col)==abs(r-row):
      conflicts+=1
  return conflicts
def get_least_conflict_row(queens,col):
  min_conflicts=float('inf')
  best_row=queens[col]
  for row in range(len(queens)):
    conflicts=count_conflicts(queens,col,row)
    if conflicts<min_conflicts:
      min_conflicts=conflicts
      best_row=row
  return best_row
def print_board(queens_list):
  n=len(queens_list)
  for r in range(n):
    row=""
    for c in range(n):
      if queens_list[c]==r:
        row+="Q "
      else:
        row+=". "
    print(row)
  print("\n")
def solve_8_queens(initial_queens,max_iterations=1000):
  n=len(initial_queens)
  queens=initial_queens[:]
  for iteration in range(max_iterations):
    conflicted_columns=[]
    for col in range(n):
      if count_conflicts(queens,col,queens[col])>0:
        conflicted_columns.append(col)
      if not conflicted_columns:
        print(f"Solution found in {iteration} iterations")
        return queens
      col=random.choice(conflicted_columns)
      new_row=get_least_conflict_row(queens,col)
      queens[col]=new_row
  print("No solution found")
  return None

n=8
initial_queens=[]
for _ in range(n):
  initial_queens.append(random.randint(0,n-1))
print(initial_queens)
print_board(initial_queens)
solution=solve_8_queens(initial_queens,100)
if solution:
  print("Solution:")
  print_board(solution)
```
