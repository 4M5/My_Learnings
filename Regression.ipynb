```python
x=[[1],[2],[3],[4],[5],[6]]
y=[[1],[4],[9],[16],[25],[36]]
import matplotlib.pyplot as plt
plt.plot(x,y)

#load model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model2=model.fit(x,y)
ynew=model2.predict(x)
print(ynew)
plt.plot(x,ynew)
print(model2.intercept_,model2.coef_)
import sklearn.metrics as m
print(m.mean_absolute_error(y,ynew))
print(m.mean_squared_error(y,ynew))
print(m.r2_score(y,ynew))
import numpy as np
print(np.sqrt(m.mean_squared_error(y,ynew)))
```

```python
#Make regression
from sklearn.datasets import make_regression
x, y = make_regression(n_samples=50, n_features=1, n_targets=1)
print(x.shape)
print(y.shape)
print(x.ndim,y.ndim)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model2=model.fit(x,y)
ynew=model2.predict(x)
print(ynew)
print(model2.intercept_,model2.coef_)
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,ynew,color="Red")
```

```python
import sklearn.metrics as m
print(m.mean_absolute_error(y,ynew))
print(m.mean_squared_error(y,ynew))
print(m.r2_score(y,ynew))
```

```python
import numpy as np
print(np.sqrt(m.mean_squared_error(y,ynew)))
```
