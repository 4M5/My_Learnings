```python
#Load data
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/content/irisexcel (2).csv")
print(data.head())
```

```python
x1=data[["sepal_length"]]
x2=data[["sepal_width"]]
plt.scatter(x1,x2,c="indigo",edgecolor='cyan',marker=',')
plt.xlabel("Sep Leng")
plt.ylabel("Sep Wid")
plt.title("IRIS")
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=5,max_iter=3)
x=data[["sepal_length","sepal_width"]]
x1=x[["sepal_length"]]
x2=x[["sepal_width"]]

model1=model.fit(x)
print(model1.labels_,"\n",model1.cluster_centers_)
plt.scatter(x1,x2,c=model.labels_,cmap='rainbow',marker="*")
```
