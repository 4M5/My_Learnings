```python
#Label Encoder
import pandas as pd
data=pd.read_excel("/content/dataset.xlsx")
datfram=pd.DataFrame(data)
data.head()
from sklearn.preprocessing import LabelEncoder
label_encoders={}
for column in datfram.columns:
  le1=LabelEncoder()
  datfram[column]=le1.fit_transform(datfram[column])
  label_encoders[column]=le1
label_encoders["Play"].classes_

x=datfram[["Outlook","Temperature","Humidity","Windy"]]
y=datfram[["Play"]]

from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier()
model2=model1.fit(x,y)
print(model2.predict([[1,1,0,0]]))

from sklearn import tree
othw=tree.export_graphviz(model2,out_file=None,feature_names=x.columns,class_names=label_encoders["Play"].classes_,filled=True,rounded=True)
import graphviz
graph = graphviz.Source(othw)
graph
```

```python
othw=tree.export_graphviz(model2,out_file=None,feature_names=x.columns,class_names=["No ","Yes"],filled=True,rounded=True)
import graphviz
graph = graphviz.Source(othw)
graph
```
