```python
import matplotlib.pyplot as plt
img1=cv2.imread('/content/download (1).jpeg')
plt.imshow(img1)
```

```python
# to avoid it (BGR to rgb)
img2=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
blue_channel=img2[:,:,0]
cv2_imshow(blue_channel)
print(blue_channel)
```

```python
#gray scale Image
gray_image=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
gray_image.shape
```

```python
#increasing brightness and contrast
b=2#increase brightness
c=50#increase contrast
adjusted= cv2.convertScaleAbs(img1,alpha=b,beta=c)
cv2_imshow(adjusted)
```

```python
adjusted= cv2.convertScaleAbs(img1,alpha=.5,beta=-50)
cv2_imshow(adjusted)
```

```python
#histogram
import numpy as np
hist1=cv2.calcHist([gray_image],[0],None,[256],[0,256])
plt.plot(hist1)
plt.show()
```

```python
#hist equal
equ=cv2.equalizeHist(gray_image)
cv2_imshow(equ)
hist2=cv2.calcHist([equ],[0],None,[256],[0,256])
plt.plot(hist2,color='black')
plt.show()
```

```python
blue_channel=img2[:,:,2]
cv2_imshow(blue_channel)
print(blue_channel)

blue_only=np.zeros_like(img2)
blue_only[:,:,2]=blue_channel
cv2_imshow(blue_only)
```
