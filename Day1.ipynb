1.Area and perimeter of rectangle

```python
l=float(input("Enter length: "))
b=float(input("Enter breadth: "))
area=l*b
perimeter=2*(l+b)
print("Area",area)
print("Perimeter",perimeter)
```

2.Check /, //, %

```python
a=int(input("Enter 1st no: "))
b=int(input("Enter 2nd no: "))
c=a/b
d=a//b
e=a%b
print("/= ",c)
print("//= ",d)
print("%= ",e)
```

3.Swapping numbers

```python
a=5
b=10
print(a,b)
temp=a
a=b
b=temp
print("After: ",a)
print("After: ",b)
```

4.Voters Eligibility

```python
age=float(input("Enter the age: "))
if(age>=18):
  print("Eligible")
else:
  print("Not Eligible")
```

5.Biggest of Two Numbers

```python
a=int(input("Enter 1st no: "))
b=int(input("Enter 2nd no: "))
if(a>b):
  print(a,"is biggest")
else:
  print(b,"is biggest")
```
