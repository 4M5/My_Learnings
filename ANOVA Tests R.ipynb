```r
#1-Way TTEST
tea<- c(10,9,6,7,8,4,6,7,4,3,6,4)
sugar = factor(rep(c("High", "Med", "Low"), each=4))
data<-data.frame(tea,sugar)
anova<- aov(tea~sugar)
anova
summary(anova)
```

```r
#Paired TTEST
prg1<- c(22,20,19,24,25,25,28,22,30,27,24,18,16,19,19,28,24,25,25,23)
prg2<- c(24,22,19,22,28,26,28,24,30,29,25,20,17,18,18,28,26,27,27,24)
ttest<-t.test(prg1,prg2,paired=TRUE)
ttest
```

```r
#ChiSquared Test
data <- matrix(c(100,40,80,30), nrow = 2, byrow = TRUE)
colnames(data) <- c("LDF", "UDF")
rownames(data) <- c("Male", "Female")
test<- chisq.test(data, correct= FALSE)
data
test
```
