```python
x=[[1],[3],[4],[6],[8]]
y=[[11],[16],[15],[20],[18]]

import matplotlib.pyplot as plt
plt.plot(x,y)
plt.scatter(x,y)


from sklearn.linear_model import LinearRegression
model=LinearRegression()

model1=model.fit(x,y)
y_new=model1.predict([[10]])
print(y_new)

y_neww=model1.predict(x)
print(y_neww)
plt.plot(x,y_neww)
plt.scatter(x,y_neww)
print(model1.intercept_,model1.coef_)
```
