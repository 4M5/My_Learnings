1> FIFO Queue

```python

from collections import deque

#1.create an empty queue
queue=deque()

#2.enqueue (add elements)
queue.append("A")
queue.append("B")
queue.append("C")
queue.append("D")
print("Queue after enque: ",list(queue))

#3.dequeue
queue.popleft()
print("Queue after Popping",list(queue))
```
